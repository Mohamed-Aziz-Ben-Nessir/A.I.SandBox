```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("/content/10k_annotated.csv")
df=df[["text","class"]]
df=df[["text","class"]]
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['class'])
df=df[["text","label"]]
train, validate, test = np.split(df.sample(frac=1, random_state=42), [int(.6*len(df)), int(.8*len(df))])
train.to_csv("train.tsv",sep="\t",index=False,header=False)
validate.to_csv("valid.tsv",sep="\t",index=False,header=False)
test.to_csv("test.tsv",sep="\t",index=False,header=False)
```
