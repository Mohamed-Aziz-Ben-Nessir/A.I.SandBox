```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
train=pd.merge(pd.read_csv("./train/client_train.csv"), pd.read_csv("train/invoice_train.csv"), on='client_id', how='right')
```

```python
train['counter_type'] = train['counter_type'].map({"ELEC":1,"GAZ":0})
train.head()
```

```python
train.columns
```

```python
X_train=train[['disrict', 'client_catg', 'region' 
       , 'tarif_type', 'counter_number',
        'counter_code', 'reading_remarque',
       'counter_coefficient', 'consommation_level_1', 'consommation_level_2',
       'consommation_level_3', 'consommation_level_4', 'old_index',
       'new_index', 'months_number']]
y_train=train["target"]
```

```python
params = {
            'objective':'binary:logistic',
            'max_depth': 4,
            'alpha': 10,
            'learning_rate': 1.0,
            'n_estimators':100
        }         
           
clf = XGBClassifier(**params)


clf.fit(X_train, y_train)
```

```python
test=pd.merge(pd.read_csv("./test/client_test.csv"), pd.read_csv("test/invoice_test.csv"), on='client_id', how='right')
```

```python
X_test=test[['disrict', 'client_catg', 'region' 
       , 'tarif_type', 'counter_number',
        'counter_code', 'reading_remarque',
       'counter_coefficient', 'consommation_level_1', 'consommation_level_2',
       'consommation_level_3', 'consommation_level_4', 'old_index',
       'new_index', 'months_number']]
```

```python
y_pred = clf.predict(X_test)
```

```python
pd.DataFrame([test["client_id"],y_pred]).to_csv("submission.csv")
```
